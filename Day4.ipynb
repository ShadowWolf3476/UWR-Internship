```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

data={
    "Age": [25, 30, None, 35],
    "Studyhours": [2, 3, None, 5],
    "Gender": ['M', 'F', 'M', 'F'],
    "Marks": [80, 85, 90, 95]
}

df = pd.DataFrame(data)
print("Original Data ")
print(df)
print(df.isnull().sum())
```

```python
print("\nData after handling missing values")
df.fillna({"Age": df["Age"].mean(), "Studyhours": df["Studyhours"].mean()}, inplace=True)
print(df)
print("\nData after handling missing values")
print(df)
```
