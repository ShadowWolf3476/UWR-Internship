```python
numbers={1,2,3,4,5}
numbers.add(6)
print(numbers)
```

# Broadcasting

```python
import numpy as np
arr=np.array([1,2,3,4,5])
arr+10
print(arr)
print(np.median(arr))
print(np.sum(arr))
print(np.mean(arr))
#exp, log,sin,,cos,tan
#min,median,sum
```
