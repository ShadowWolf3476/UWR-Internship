# Questions

```python
n = int(input("Enter the number: "))
if n>0:
    print(n,"is positive")
elif n<0:
    print(n,"is negative")
else:
    print(n,"is zero")
```

```python
yr=int(input("Enter the year: "))
if (yr%4==0 and yr%100!=0) or (yr%400==0):
    print(yr,"is a leap year")
else:
    print(yr,"is not a leap year")
```

```python
sm=0
for i in range(1,101):
    sm+=i
print("Sum of numbers from 1 to 100 is:", sm)
```

```python
wrd=input("Enter the word: ")
vwl=0
for i in wrd:
    if i in "aeiouAEIOU":
        vwl+=1
print("Total vowels in the word are:", vwl)
```

```python
n=5
while n!=0:
    n=int(input("Enter the number: "))
print("You entered zero. Loop terminated.")
```

```python
n=int(input("Enter the number: "))
temp=n
rev=0
while temp>0:
    dig=temp%10
    rev=rev*10+dig
    temp//=10
print("The reverse of the number is:", rev)
```

```python
n=int(input("Enter the number: "))
if n%3==0:
    print("fizz")
elif n%5==0:
    print("buzz")
```

```python
n=int(input("Enter the number: "))
if n>1:
    for i in range(2,n):
        for j in range(2,i):
            if i%j==0:
                break
        else:
            print(i)
```
